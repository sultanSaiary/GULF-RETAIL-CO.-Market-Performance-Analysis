```python
sales_transactions_df['Date'] = pd.to_datetime(sales_transactions_df['Date'], errors='coerce', dayfirst=True)
store_complaints_df['Date'] = pd.to_datetime(store_complaints_df['Date'], errors='coerce', dayfirst=True)
customer_data_df['Registration_Date'] = pd.to_datetime(customer_data_df['Registration_Date'], errors='coerce', dayfirst=True)
store_information_df['Opening_Date'] = pd.to_datetime(store_information_df['Opening_Date'], errors='coerce', dayfirst=True)
marketing_campaigns_df['Month'] = pd.to_datetime(marketing_campaigns_df['Month'], errors='coerce', dayfirst=True)

# التأكد من أن العمود أصبح من نوع تاريخ
print(sales_transactions_df['Date'].dtype)
```

```python
# دمج بيانات المبيعات مع معلومات المتاجر
merged_df = pd.merge(sales_transactions_df, store_information_df, on='Store_ID', how='inner')

# عرض أول 5 صفوف بعد الدمج
print("Merged Data Sample:")
merged_df.head()
```

```python
pd.options.display.float_format = '{:,.0f}'.format
```

```python
# تحليل الأداء الإقليمي
regional_performance = merged_df.groupby('Region').agg(
    total_revenue=('Revenue', 'sum'),
    transaction_count=('Transaction_ID', 'count')
)
regional_performance = regional_performance.sort_values(by='total_revenue', ascending=False)

# عرض نتائج الأداء الإقليمي
print("Regional Performance Metrics:")
print(regional_performance)
```

```python
# حساب معدل النمو الشهري للإيرادات
merged_df['Date'] = pd.to_datetime(merged_df['Date'])
monthly_revenue = merged_df.groupby([merged_df['Date'].dt.to_period('M'), 'Region']).agg(
    monthly_revenue=('Revenue', 'sum')
).reset_index()

monthly_revenue_pivot = monthly_revenue.pivot_table(
    values='monthly_revenue', 
    index='Date', 
    columns='Region', 
    aggfunc='sum'
)

monthly_revenue_pivot_pct_change = monthly_revenue_pivot.pct_change().fillna(0) * 100

# عرض أول 18 شهر  ل معدل النمو الشهري
monthly_revenue_pivot_pct_change.head(18)
```

```python
# 1. Bar chart for total revenue by region
plt.figure(figsize=(10,6))
sns.barplot(x=regional_performance.index, y='total_revenue', data=regional_performance, palette='viridis')
plt.title('Total Revenue by Region')
plt.xlabel('Region')
plt.ylabel('Total Revenue (SAR)')
plt.xticks(rotation=45)
plt.show()

# 2. Line chart for monthly revenue trend per region
plt.figure(figsize=(12,6))
for region in monthly_revenue_pivot.columns:
    plt.plot(monthly_revenue_pivot.index.astype(str), monthly_revenue_pivot[region], label=region)

plt.title('Monthly Revenue Trend by Region')
plt.xlabel('Month')
plt.ylabel('Revenue (SAR)')
plt.xticks(rotation=45)
plt.legend(title='Region')
plt.show()
```

```python
# الخطوة 1: فلترة المبيعات الموجبة فقط وإضافة عمود الشهر
sales_clean = sales_transactions_df[sales_transactions_df['Revenue'] > 0].copy()
sales_clean['Month'] = pd.to_datetime(sales_clean['Date']).dt.to_period('M').astype(str)

# تجميع الإيرادات لكل فرع في كل شهر
sales_monthly = sales_clean.groupby(['Store_ID', 'Month'])['Revenue'].sum().reset_index()
sales_monthly.columns = ['Store_ID', 'Month', 'Monthly_Revenue']

print("عدد الصفوف:", len(sales_monthly))
print(sales_monthly.head())
```

```python
# الخطوة 2: توحيد عمود الشهر في ملف التسويق
marketing_clean = marketing_campaigns_df.copy()
marketing_clean['Month'] = pd.to_datetime(marketing_clean['Month']).dt.to_period('M').astype(str)

print("عدد الصفوف:", len(marketing_clean))
print(marketing_clean.head(20))
```

```python
# الخطوة 3: دمج المبيعات مع التسويق عن طريق Store_ID + Month
merged_ddf = pd.merge(sales_monthly, marketing_clean, on=['Store_ID', 'Month'], how='inner')

print("عدد الصفوف بعد الدمج:", len(merged_df))
print(merged_ddf.head())
```

```python
# الخطوة 4: حساب ROI
# ROI = (الإيراد - الميزانية) ÷ الميزانية
merged_ddf['ROI'] = (merged_ddf['Monthly_Revenue'] - merged_ddf['Budget_SAR']) / merged_ddf['Budget_SAR']

print(merged_ddf[['Store_ID', 'Month', 'Monthly_Revenue', 'Budget_SAR', 'ROI']].head(10))
```

```python
# الخطوة 5: إضافة المنطقة من ملف الفروع
merged_with_region = pd.merge(
    merged_ddf,
    store_information_df[['Store_ID', 'Region']],
    on='Store_ID',
    how='left'
)

print(merged_with_region[['Store_ID', 'Region', 'ROI']].head())
```

```python
# الخطوة 6: متوسط ROI لكل منطقة
roi_by_region = merged_with_region.groupby('Region')['ROI'].mean().reset_index()
roi_by_region.columns = ['Region', 'Average_ROI']
roi_by_region = roi_by_region.sort_values('Average_ROI', ascending=False)

fig1 = px.bar(
    roi_by_region,
    x='Region',
    y='Average_ROI',
    title='Average Marketing ROI by Region',
    color='Average_ROI',
    color_continuous_scale='Greens',
    text=roi_by_region['Average_ROI'].apply(lambda x: f'{x:.1f}x')
)
fig1.update_traces(textposition='outside')
fig1.update_layout(showlegend=False)
fig1.show()
```

```python
# الخطوة 7: متوسط ROI لكل نوع حملة
roi_by_campaign = merged_with_region.groupby('Campaign_Type')['ROI'].mean().reset_index()
roi_by_campaign.columns = ['Campaign_Type', 'Average_ROI']
roi_by_campaign = roi_by_campaign.sort_values('Average_ROI', ascending=False)

fig2 = px.bar(
    roi_by_campaign,
    x='Campaign_Type',
    y='Average_ROI',
    title='Average Marketing ROI by Campaign Type',
    color='Average_ROI',
    color_continuous_scale='Blues',
    text=roi_by_campaign['Average_ROI'].apply(lambda x: f'{x:.1f}x')
)
fig2.update_traces(textposition='outside')
fig2.update_layout(showlegend=False)
fig2.show()
```

```python
# الخطوة 8: Scatter Plot - الميزانية مقابل الإيراد
fig3 = px.scatter(
    merged_with_region,
    x='Budget_SAR',
    y='Monthly_Revenue',
    color='Campaign_Type',
    hover_data=['Store_ID', 'Region', 'Month'],
    title='Marketing Budget vs Revenue by Campaign Type',
    labels={
        'Budget_SAR': 'Budget (SAR)',
        'Monthly_Revenue': 'Revenue (SAR)'
    }
)
fig3.show()
```
