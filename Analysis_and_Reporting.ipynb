```python
# ===================================
# رسم التوقع لكل مدينة - Bar Chart
# ===================================

fig1 = px.bar(
    forecast_df,
    x='City',
    y='Q3_2024_Forecast',
    title='Forecasted Revenue by City — Q3 2024',
    color='Q3_2024_Forecast',
    color_continuous_scale='Blues',
    text=forecast_df['Q3_2024_Forecast'].apply(lambda x: f'{x:,.0f} SAR')
)
fig1.update_traces(textposition='outside')
fig1.update_layout(
    showlegend=False,
    xaxis_title='City',
    yaxis_title='Forecasted Revenue (SAR)'
)
fig1.show()
```
